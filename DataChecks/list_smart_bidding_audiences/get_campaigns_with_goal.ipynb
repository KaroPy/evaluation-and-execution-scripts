```python
import logging
import pandas as pd
from general_functions.return_workspace_ids import return_workspace_ids
from general_functions.constants import return_api_url
from general_functions.call_api_with_account_id import send_to_innkeepr_api_paginated, call_api_with_accountId
```

```python
url = return_api_url()
print(f"url = {url}")
workspaces = return_workspace_ids(tracking_started=False)
valid_source = "googleAdwords"
```

```python
conversion_signals = pd.read_csv("DataChecks/list_smart_bidding_audiences/conversion_actions.csv")
conversion_signals.head()
```

```python
for workspace in conversion_signals["workspace"].unique():
    print(f"Workspace: {workspace}")
    workspace_id = [entry["id"] for entry in workspaces if entry["name"] == workspace][0]
    sources = call_api_with_accountId(
        endpoint_url=f"{url}/sources/query",
        accountID=workspace_id,
        content={"name": valid_source},
        logger=logging
    )
    print(f"Sources: {sources}")

    treatments = send_to_innkeepr_api_paginated(
        api_url=f"{url}/treatments/query",
        accountID=workspace_id,
        content={"source": sources[0]["id"]},
        logger=logging
    )
    treatments = pd.json_normalize(treatments)
    print(f"Treatments: {len(treatments)}")
    break
    
```
