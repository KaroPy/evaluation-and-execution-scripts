```python
import os
import json
import pandas as pd
from general_functions.conncet_s3 import S3Connection
from general_functions.return_workspace_ids import return_workspace_ids
from general_functions.constants import return_api_url
```

```python
url = return_api_url()
list_workspaces = return_workspace_ids()
tracking_started = []
s3client = S3Connection()
path = "firehose/events/"
```

```python
for workspace in list_workspaces:
    workspace_id = workspace["id"]
    workspace_name = workspace["name"]
    print(f"Get starting date for workspace: {workspace_name} ({workspace_id})")
    files = s3client.list_files(workspace_id, path, delimiter="/")
    dates = [name.split("/")[2] for name in files]
    print(f"Dates: {dates}")
    if len(dates) == 0:
        min_date = None
        max_date = None
    else:
        min_date = min(dates)
        max_date = max(dates)
    tracking_started.append({
        "id": workspace_id,"workspace": workspace_name,"starting_date": min_date,"most_recent_date": max_date})
tracking_started
```

```python
with open('/Users/karolinegriesbach/Documents/Innkeepr/Git/evaluation-and-execution-scripts/DataChecks/accounts_with_tracking/workspaces_starting_dates.json', 'w') as f:
    json.dump(tracking_started, f, indent=4)
```
