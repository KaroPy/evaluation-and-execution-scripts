```python
import pandas as pd
from datetime import datetime
```

# Define Variables

```python
path_to_data = "/Users/karolinegriesbach/Documents/Innkeepr/Git/consumption-based-costs/data/2024-01-01_to_2025-06-02/2025-06-03 19:18:08.255219/final_costs_with_azure_and_aws_and_db_2024-01-01_2025-06-02.csv"
expected_cost = {
    "azure": {"2025-05":212,"2025-04":210},
    "stackit": {"2025-05":2336.94,"2025-04":2394.9},
    "aws": {"2025-05":3940.65,"2025-04":3996.67},
    "databricks": {"2025-05":1431.79,"2025-04":1152.16}
}
path_to_save = f"data/{datetime.today().date()}"
path_to_save
```

```python
import os
os.makedirs(path_to_save, exist_ok=True)
```

# Load data

```python
df = pd.read_csv(path_to_data)
df = df[[col for col in df.columns if "Unnamed" not in col]]
df
```

# Check costs of the last april 2025 and may 2025 and compare with expected data

```python
min_date = "2025-04-01"
max_date = "2025-05-31"
costs_apil_may = df[(df["date"]>=min_date)&(df["date"]<=max_date)]
print(f"Daterange: {costs_apil_may['date'].min()} - {costs_apil_may['date'].max()} ")
```

```python
costs_apil_may["month"] = pd.to_datetime(costs_apil_may["date"], format="mixed")
costs_apil_may["month"] = pd.to_datetime(costs_apil_may["date"], format="mixed", utc=True).dt.to_period("M")
costs_apil_may_grouped = costs_apil_may.groupby(by=["cloud","month"])["charge"].sum()
costs_apil_may_grouped = pd.DataFrame(costs_apil_may_grouped).reset_index()
costs_apil_may_grouped["month"] = costs_apil_may_grouped["month"].astype("str")
costs_apil_may_grouped
```

```python
costs_apil_may_grouped["expected_costs"] = costs_apil_may[["cloud","month"]].apply(lambda x: expected_cost[x["cloud"]][x["month"]], axis=1)
costs_apil_may_grouped
```

## Test just aws

```python
aws = costs_apil_may[costs_apil_may["cloud"]=="aws"]
aws
```

```python
aws_by_serviceCategoryName = aws.groupby(by=["month","serviceCategoryName"])["charge"].sum()
aws_by_serviceCategoryName = pd.DataFrame(aws_by_serviceCategoryName).reset_index()
aws_by_serviceCategoryName
```

```python
aws_by_serviceCategoryName.to_csv(f"{path_to_save}/aws_costs_april_may.csv")
```
