```python
# Die Variable „df“ wurde aus dem URI geladen: /Users/karolinegriesbach/Documents/Innkeepr/Git/evaluation-and-execution-scripts/SprintStories/EN-2933-Fix-Loading-Lstm/data/temp/2025-11-01_2025-11-19_targeting_stats_duratiaon_0.csv
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

start_date = "2025-11-01"
end_date = "2025-11-19"
duration = None
include_workspaces = ["Tchibo", "ESN", "More"]
df = pd.read_csv(f'/Users/karolinegriesbach/Documents/Innkeepr/Git/evaluation-and-execution-scripts/SprintStories/EN-2933-Fix-Loading-Lstm/data/temp/{start_date}_{end_date}_targeting_stats_duratiaon_{duration}.csv')
df = df[df["account"].isin(include_workspaces)]
df["account"].value_counts()
```

```python
df.groupby(by=["date","account"])["audience_id"].count()
```

```python
average_run_time_per_day = df.groupby(by=["date","audience_id","account"])["total_run_time"].mean()
average_run_time_per_day = average_run_time_per_day.reset_index().sort_values(by="date").rename(columns={"total_run_time" : "M(total_run_time)"})
average_run_time_per_day["M(run_time in min)"] = average_run_time_per_day["M(total_run_time)"]/60
average_run_time_per_day
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)
sns.lineplot(data=average_run_time_per_day, x="date", y="M(run_time in min)", hue="account", style="account", markers="o", ax=ax)
plt.grid(True)
plt.xticks(rotation=90)
```
